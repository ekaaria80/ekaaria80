```python
def month (series):
    if series == "january":
        return 0
    if series == "february":
        return 1
    if series == "march":
        return 2
    if series == "april":
        return 3
    if series == "may":
        return 4
    if series == "june":
        return 5
    if series == "july":
        return 6
    if series == "august":
        return 7
    if series == "september":
        return 8
    if series == "october":
        return 9
    if series == "november":
        return 10
    if series == "december":
        return 11
```
